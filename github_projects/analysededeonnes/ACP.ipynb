```python
cols=['highway-mpg','engine-size','horsepower','curb-weight']
matrix=df[cols].to_numpy()
print(matrix)
```

# Calcule du matrice centre reduite :

```python
matrix_mean = np.mean(matrix, axis=0)
matrix_std = np.std(matrix, axis=0)
matrixCentreReduite = (matrix- matrix_mean) / matrix_std
print(matrixCentreReduite)
```

## Calculer des valeurs propres et les vecteurs propres :

```python
U, s, Vt = np.linalg.svd(matrix, full_matrices=False)

valeur_propre = s ** 2 / (matrix.shape[0] - 1)
vecteur_propre =  U.T
```

```python
# Afficher les résultats
print("Valeurs propres :", valeur_propre)
print("Vecteurs propres :\n", vecteur_propre)
t1=np.transpose(vecteur_propre)
P1=matrixCentreReduite@t1[0,:]
print("la premiere composante:",P1)
P2=matrixCentreReduite@t1[1,:]
print("la deuxieme composante:",P2)
P2
```
